```python
train_path = './data/weiboNER_2nd_conll.train'
train_data = []
with open(train_path, 'r', encoding = 'utf-8') as f:
    tmp = []
    for line in f:
        if line != '\n': 
            tmp.append(line.strip('\n'))
        else:
            train_data.append(tmp)
            tmp = []
```

```python
train = []
for data in train_data:
    sentence = []
    marks = []
    for token in data:
        ch, mark = token.split('\t')
        ch = ch[0]
        sentence.append(ch)
        marks.append(mark)
    train.append([sentence, marks])
len(train)
```

```python
tag_set = set()
for data in train:
    tag_set.update(data[1])
len(tag_set)
```

```python
entity_types = ['GPE.NAM',
 'GPE.NOM',
 'LOC.NAM',
 'LOC.NOM',
 'ORG.NAM',
 'ORG.NOM',
 'PER.NAM',
 'PER.NOM']
```

```python
def get_entity(data, label):
    begins = []
    ends = []
    for index in range(len(data[1])):
        if data[1][index].startswith('B-'+label):
            begins.append(index)
    for begin in begins:
        if begin == len(data[0]):
            ends.append(begin)
        else: 
            for i in range(begin+1, len(data[1])):
                if not data[1][i].startswith('I-'+label):
                    ends.append(i-1)
                    break
            
    return begins,ends
```

```python
def build_data(d):
    data = {}
    data['text'] = d[0]
    for entity_type in entity_types:
        data[entity_type+'-index'] = get_entity(d, entity_type)
    return data
build_data(train[10])
```

```python
final = {}
for index in range(len(train)):
    final[index] = build_data(train[index])
```

```python
def counter(label, data):
    count = 0
    entity = []
    for i in range(len(data)):
        count += len(data[i][label+'-index'][0])
        for begin, end in zip(data[i][label+'-index'][0],data[i][label+'-index'][1]):
            entity.append(''.join(data[i]['text'][begin:end+1]))
    return entity,count
counter('PER.NOM', final)
```
